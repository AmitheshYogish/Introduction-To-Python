### QUESTION 3
Write a code to create a Triangle class that gets three side-lengths of the triangle as x ,y and z as parameters. Check whether the triangle is valid or not ,using parameters and print the type of the triangle. Also print the area of the triangle.

Is_valid(): Returns valid if triangle is valid otherwise returns Invalid

A triangle is valid when the sum of its two side-length are greater than the third one. That means the triangle is valid if all three condition are satisfied:

•	a+ b > c

•	a + c > b

•	b + c > a

Sample Input:

5
6
8

Sample Output:

Valid
Scalene
Obtuse
14.981238266578634





```python
class Triangle:
    def __init__(self,a,b,c):
        self.a= a
        self.b= b
        self.c= c
    def Is_valid(self):
        valid=False
        if self.a+self.b>self.c:
            if self.b+self.c>self.a:
                if self.a+self.c>self.b:
                    valid=True
        if valid==True:
            return("Valid")
        else:
            return("Invalid")
    def Angle_Classification(self):
        if Triangle.Is_valid(self)=="Invalid":
           return("Invalid")
        else:
            x=min(self.a,self.b,self.c)
            y=max(self.a,self.b,self.c)
            z=(self.a+self.b+self.c)-(x+y)
            if x**2+z**2>y**2:
                return("Acute")
            elif x**2+z**2==y**2:
                return("Right")
            else:
                return("Obtuse")
    def Side_Classification(self):
        x=self.a
        y=self.b
        z=self.c
        if Triangle.Is_valid(self)=="Invalid":
           return("Invalid")
        else:
          if (x==y)and (x==z):
            return("Equilateral")
          elif ((x==y)and (x!=z)) or ((x==z)and(x!=z)) or ((y==z)and(y!=x)):
            return("Isosceles")
          else:
            return("Scalene")
    def Area(self):
        if Triangle.Is_valid(self)=="Invalid":
           return("Invalid")
        else:
            s=(self.a+self.b+self.c)/2
            area=(s*(s - self.a)*(s- self.b)*(s- self.c))**0.5
            return(area)

a=int(input())
b=int(input())
c=int(input())
T = Triangle(a,b,c)
print(T.Is_valid())
print(T.Side_Classification())
print(T.Angle_Classification())
print(T.Area())
```

1.Second_to_Minutes that converts the value of time into minutes and returns the output in minutes min second sec format.

For example: if seconds is 130, the method should return 2 min 50sec. 

2.Seconds_to_Hours that converts the value of time into hours and returns the output in hours hr minutes min and seconds sec format.

For example: if seconds is 130, the method should return 0hr 2 min 50 sec

3.Second_to_Days that converts the value of time into days and returns the output in days days hours hr minutes min and seconds sec format. 

For example: if seconds is 130, the method should return 0 days 0 hr 2 min 50 sec 

Sample Input:

a= TimeConverter (35)

Sample Output:

0 min 35 sec

0 hr 0 min 35 sec

0 days 0 hr 0 min 35 sec


```python
class TimeConverter:
    def __init__(self,time):
        self.time=time
    def Second_to_Minutes(self):
        m = self.time//60
        s = self.time % 60
        return (str(m)+' min '+str(s)+' sec')
    def Second_to_Hours(self):
        m = self.time//60
        s = self.time % 60
        h = m // 60
        m = m % 60
        return (str(h)+' hr '+str(m)+' min '+str(s)+' sec')
    def Second_to_Days(self):
        m = self.time//60
        s = self.time % 60
        h = m // 60
        m = m % 60
        d = h // 24
        h = h % 24
        return (str(d)+' days '+str(h)+' hr '+str(m)+' min '+str(s)+' sec')
sec = int(input())
a = TimeConverter(sec)
print(a.Second_to_Minutes())
print(a.Second_to_Hours())
print(a.Second_to_Days())
```

### QUESTION 9
A Human moves on a plane surface starting from the origin (0,0). The Human can move toward UP, DOWN, LEFT and RIGHT with a given steps. Write a program to compute the distance from current position after a sequence of movement and original point. If the distance is a float, then just print the nearest integer.

Note:

UP 3

DOWN 5 

RIGHT 4

LEFT 2  
(The numbers after the direction are steps.)

Sample Input:

UP 3

DOWN 5 

RIGHT 4

LEFT 2

Sample Output:

3



```python
import math
pos = [0,0]
while True:
    s = input()
    if not s:
        break
    movement = s.split(" ")
    direction = movement[0]
    steps = int(movement[1])
    if direction=="UP":
        pos[0]+=steps
    elif direction=="DOWN":
        pos[0]-=steps
    elif direction=="LEFT":
        pos[1]-=steps
    elif direction=="RIGHT":
        pos[1]+=steps
    else:
        pass

print (int(round(math.sqrt(pos[1]**2+pos[0]**2))))
```

### QUESTION 14
Write a python program to get a list from the user

1.use reduce to compute the sum of the list using lamda function

2.use reduce to compute maximum element from the list

Sample Input:

2 3 4 5 6 7 9

Sample Output:

The sum of the list elements is : 36

The maximum element of the list is : 9


```python
import functools
 
# initializing list
x = input().split()
lis = [int(i) for i in x]
 
# using reduce to compute sum of list
print("The sum of the list elements is : ", end="")
print(functools.reduce(lambda a, b: a+b, lis))
 
# using reduce to compute maximum element from list
print("The maximum element of the list is : ", end="")
print(functools.reduce(lambda a, b: a if a > b else b, lis))
```

### QUESTION 15
Write a Python program to create all possible permutations from a given collection of distinct numbers.

Sample Input:

1 2 3

Sample Output:

Original Collection:  [1, 2, 3]

Collection of distinct numbers:
 [[3, 2, 1], [2, 3, 1], [2, 1, 3], [3, 1, 2], [1, 3, 2], [1, 2, 3]]

```python
def permute(nums):
  result_perms = [[]]
  for n in nums:
    new_perms = []
    for perm in result_perms:
      for i in range(len(perm)+1):
        new_perms.append(perm[:i] + [n] + perm[i:])
        result_perms = new_perms
  return result_perms

my_nums = input().split()
my_nums = [int(x) for x in my_nums]

print("Original Collection: ",my_nums)
print("Collection of distinct numbers:\n",permute(my_nums))
```
